```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# 1. Load the pre-trained DistilBERT model and its corresponding tokenizer
# This model is specifically fine-tuned for sentiment analysis (SST-2 dataset)
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# 2. Define the input text samples
test_sentences =  [
    "The bidirectional training of BERT allows for a much deeper understanding of language context.",
    "The model failed to converge, and the training process was inefficient and slow."
]

# 3. Tokenize the inputs
# BERT requires inputs to be converted into numerical tensors [cite: 29, 30]
inputs = tokenizer(test_sentences, padding=True, truncation=True, return_tensors="pt")

# 4. Perform inference
with torch.no_grad():
    outputs = model(**inputs)

# 5. Process the raw output (logits) into probabilities using Softmax
logits = outputs.logits
probabilities = torch.nn.functional.softmax(logits, dim=-1)

# 6. Determine the predicted class (0 for NEGATIVE, 1 for POSITIVE)
predicted_class_ids = torch.argmax(probabilities, dim=-1).tolist()
labels = ["NEGATIVE", "POSITIVE"]

# 7. Display Results
print("--- Sentiment Analysis Results ---")
for i, text in enumerate(test_sentences):
    label = labels[predicted_class_ids[i]]
    conf = probabilities[i][predicted_class_ids[i]].item()
    print(f"Text: '{text}'")
    print(f"Prediction: {label} (Confidence: {conf:.2%})\n")
```
